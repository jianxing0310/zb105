```python
import jieba
from pymongo import MongoClient 
import time
tStart=time.time() #計算所需時間(計時開始)

jieba.set_dictionary('E:/text_mining/dict.txt')  #切換至中文繁體字庫
jieba.load_userdict('E:/text_mining/dict_keyw.txt')  #
jieba.load_userdict('E:/text_mining/dict_cbdic.txt')  #

client = MongoClient('10.120.28.14',27017)
database = client['test']
collection =database['test5']


negative_set=set()
positive_set=set()

negative_write=set()
positive_write=set()

f=open('E:/text_mining/negative.txt','r')
for word in f.readlines():
    word=word.strip()
    if word not in negative_set:
        negative_set.add(word)
f.close()


f=open('E:/text_mining/positive.txt','r')
for word in f.readlines():
    word=word.strip()
    if word not in positive_set:
        positive_set.add(word)
f.close()

a=0
b=0
for post in collection.find(): 
    summary = post['content']
    content=jieba.cut(summary)
    for word in content:
        #print word
        if word.encode('utf-8') in negative_set:
            a=a+1
            negative_write.add(word.encode('utf-8'))
        if word.encode('utf-8') in positive_set:
            b=b+1
            positive_write.add(word.encode('utf-8'))
    break
print '負向', a
print '正向', b

negative=open('negative_word.txt','w')
for word in negative_write:
    negative.write(word+'\n')
negative.close()

positive=open('positive_word.txt','w')
for word in positive_write:
    positive.write(word+'\n')
positive.close()

tEnd = time.time() #記時(時間結束)
  
print ' 花了 {} s'.format(tEnd-tStart)    #將經過時間print 出來
```
