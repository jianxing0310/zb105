# 有無index

```python
from pymongo import MongoClient
client = MongoClient('10.120.28.14', 27017)
database = client["qoo"]
collection = database["test1"]
```

```python
tfidf_set = set([])
import time
start = time.clock()
resp = collection.find({"$and":[
                        {"tfidf":"鄭捷"},
                        {"tfidf":"江子翠"},    
                            ]},
                    {"date":1,"_id":0,"title":1}).sort("date",-1)
#print resp.count()
end = time.clock() - start
print end
#print resp.explain()
for ele in resp:
    tfidf_set.add(ele["title"])  # 將標題放到set中
```
